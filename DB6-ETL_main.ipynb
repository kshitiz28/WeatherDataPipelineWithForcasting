## FactHourly Table

```
def generate_fact_hourly_with_log():
    upsert_into_facthourly()
    df_fact_weather_hourly = spark.read.table("fact_weather_data_hourly")
    get_fact_with_log('FACT', 'fact_weather_hourly', df_fact_weather_hourly)
```

## FactDaily Table

```
def generate_fact_daily_with_log():
    insert_into_factdaily()
    df_fact_weather_daily = spark.read.table("fact_weather_data_daily")
    get_fact_daily_with_log('FACT', 'fact_weather_daily', df_fact_weather_daily)
```

## Pipeline loop

```
def overall_pipeline():
    # Get the city data from dim_city
    df= spark.read.format("delta").table('dim_city')
    df = df.withColumn('id', col('id').cast('int'))
    df_raw = get_weather('RAW', 'raw_weather', df) # api-to-raw
    get_process('PROCESSED', 'cleansed_weather', df_raw) #raw-to-cleansed
    generate_fact_hourly_with_log() #FactHourlyTable
    generate_fact_daily_with_log() #FactDailyTable
```

```
overall_pipeline()
```
